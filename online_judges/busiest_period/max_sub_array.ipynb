```python
class Solution(object):
    
    def find_max_subarray(self, array):
        if array is None: 
            raise TypeError("array can't be None")
        if len(array) == 0:
            return None
        
        max_sum = array[0]
        max_s = 0
        max_e = 0
        cur_sum = array[0]
        cur_s = 0
        cur_e = 0
    
        
        for i in range(1, len(array)):
            print("before: i = {}, cur = ({}, {}), {} max = ({}, {}), {}".format(i, cur_s, cur_e, cur_sum, max_s, max_e, max_sum))
            cur_sum += array[i]
            if cur_sum > 0:
                cur_e = i
                if cur_sum - array[cur_s] > cur_sum:
                    cur_sum -= array[cur_s]
                    cur_s += 1
            else:
                cur_s = i
                cur_e = i
                cur_sum = array[i]
            if cur_sum > max_sum:
                max_s = cur_s
                max_e = cur_e
                max_sum = cur_sum
            print("after:  i = {}, cur = ({}, {}), {} max = ({}, {}), {}".format(i, cur_s, cur_e, cur_sum, max_s, max_e, max_sum))

        return max_s, max_e
        
        
```

```python
import unittest

class TestSolution(unittest.TestCase):

    def test_max_sub_array(self):
        solution = Solution()
        self.assertRaises(TypeError, solution.find_max_subarray, None)
        self.assertEqual(solution.find_max_subarray([]), None)
        self.assertEqual(solution.find_max_subarray([1, 2, 3, 4, -10, 5]), (0, 3))
        self.assertEqual(solution.find_max_subarray([1, 2, 3, 4, -10, 11]), (5, 5))
        self.assertEqual(solution.find_max_subarray([-5, -3, -1, -2, -6]), (2, 2))        
        self.assertEqual(solution.find_max_subarray([-5, -4, -1, 1, -6]), (3, 3))        
        print('Success: test_find_busiest_period')


def main():
    test = TestSolution()
    test.test_max_sub_array()


if __name__ == '__main__':
    main()
```
